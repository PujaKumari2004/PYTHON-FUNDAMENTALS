```python
# Want to  check rating 
Rating = df_final.groupby(['Aggregate rating', 'Rating color', 'Rating text']).size().reset_index().rename(columns={0:'Rating count'})
Rating
```

# Observations
1. When the Rating are between the 4.5 and 4.9 is indicate the food is Excellent.
2. When the Rating are between the 4.0 and 4.4 is indicate the food is Very Good.
3. When the Rating are between the 3.5 and 3.9 is indicate the food is Good.
4. When the Rating are between the 2.5 and 3.4 is indicate the food is Average.
5. When the Rating are between the 1.8 and 2.4 is indicate the food is Poor.

```python
Rating.head(5)
```

```python
# through plt.rcparams we change the whole  size of the grapgh 
plt.rcParams['figure.figsize'] = (12,6)
sns.barplot(data = Rating, x="Aggregate rating", y = "Rating count")
```

```python
sns.barplot(data = Rating, x="Aggregate rating", y = "Rating count" ,hue = "Rating color" ,palette=["blue","red","orange","yellow","green","green"])
```

# OBSERVATION 
1) Not Rating count is very high.
2) Maximum number of rating are between 2.5  and 3.4.
3) 3.1 and 3.2 have maximum rating count.

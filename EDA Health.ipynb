```python
#Counting Courses which are opted by most of the student

course_count = df["What is your course?"].value_counts().head()
course_count
```

```python
#Counting number of Female and Male

count_gender = df["Choose your gender"].value_counts()
count_gender
```

```python
#Finding Mean Age of Female and Male

f_meanage = df["Age"][df["Choose your gender"] == "Female"].mean()

m_meanage = df["Age"][df["Choose your gender"] == "Male"].mean()

print("Mean Age of Female Students:", f_meanage)
print("Mean Age of Male Students:", m_meanage)
```

```python
#Finding Number of Students who are married and not married respectively

marrital_status = df["Marital status"].value_counts()
marrital_status
```

```python
#Percentage of students facing depression

depression = df[df["Do you have Depression?"] == "Yes"]["Choose your gender"].count()
depr_perc = depression / len(df["Do you have Depression?"]) * 100
print(depr_perc)
```

```python
#Percentage of students facing anxiety

anxiety = df[df["Do you have Anxiety?"] == "Yes"]["Choose your gender"].count()
anx_perc = anxiety / len(df["Do you have Anxiety?"]) * 100
print(anx_perc)
```

```python
#percentage of student having panic attacks

panic_attacks = df[df["Do you have Panic attack?"] == "Yes"]["Choose your gender"].count()
panic_perc = panic_attacks / len(df["Do you have Panic attack?"]) * 100
print(panic_perc)
```
